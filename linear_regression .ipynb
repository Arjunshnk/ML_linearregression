```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
x = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.scatter(x,y,alpha=0.3)
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.xlabel('production_budget')
plt.ylabel('worldwidegross')
plt.title('film cost vs revenue cost')
```

```python
regression = LinearRegression()
```

```python
regression.fit(x,y)
```

```python
regression.coef_ # theta value
```

```python
regression.intercept_ # x and y intercepts of the graph
```

```python
plt.scatter(x,y,alpha=0.25)
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.ylabel('worldwidegross')
plt.xlabel('production_budget')
plt.title('film cost vs revenue cost',color='red')
plt.plot(x,regression.predict(x),color='black',linewidth=3)
```
